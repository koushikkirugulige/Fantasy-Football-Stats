```python
import pandas as pd
import urllib.request
#change The URL and file name and run all cells and voilà
UrlgoalandShot = 'https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F12%2Fgca%2FLa-Liga-Stats&div=div_stats_gca'
UrlDefence = 'https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F12%2Fdefense%2FLa-Liga-Stats&div=div_stats_defense'
UrlStdStats = 'https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F12%2Fstats%2FLa-Liga-Stats&div=div_stats_standard'
UrlPassStats = 'https://widgets.sports-reference.com/wg.fcgi?css=1&site=fb&url=%2Fen%2Fcomps%2F12%2Fpassing%2FLa-Liga-Stats&div=div_stats_passing'
file = 'laligaAllStats.csv'
```

```python
def read_fbref(url,colslist):
    df = pd.read_html(url,encoding = 'utf_8')[0]
    df = df[colslist].copy()
    #display(df)
    df.columns = df.columns.droplevel()
    #display(df)
    df.drop_duplicates(inplace=True)
    df = df[df.Player != 'Player']
    display(df.columns)
    display(df.head())
    return df
```

```python
lst = ['Unnamed: 0_level_0','Unnamed: 1_level_0','Unnamed: 3_level_0','Unnamed: 4_level_0','Unnamed: 7_level_0','SCA','SCA Types','GCA']
df_attack = read_fbref(UrlgoalandShot,lst)
df_attack = df_attack[['Player', 'Pos', 'Squad','90s','SCA90','Drib', 'Fld','GCA90']]
df_attack.rename(columns={'Pos':'Position','SCA90':'Shot-Creating Actions90','Drib':'Successful dribbles','Fld':'Fouls Drawn','GCA90':'Goal-Creating Actions per 90 minutes'},inplace=True)
```

```python
lst = ['Unnamed: 0_level_0','Unnamed: 1_level_0','Unnamed: 3_level_0','Unnamed: 4_level_0','Unnamed: 7_level_0','Tackles','Vs Dribbles','Pressures','Blocks','Unnamed: 30_level_0']
df_def = read_fbref(UrlDefence,lst)
df_def = df_def[['Player', 'Pos', 'Squad','90s','TklW','Past','%','Sh','Pass','Err']]
df_def.rename(columns={'Pos':'Position','TklW':'Tackles Won','Past':'Dribbled Past','%':'Successful Pressure%','Sh':'Shots Blocked','Pass':'Pass Blocked','Err':'Errors led to Shot'},inplace=True)
```

```python
lst = ['Unnamed: 0_level_0','Unnamed: 1_level_0','Unnamed: 3_level_0','Unnamed: 4_level_0','Playing Time','Per 90 Minutes']
df_std_stats = read_fbref(UrlStdStats,lst)
df_std_stats =df_std_stats[['Player','Pos', 'Squad','Starts','Gls','Ast', 'G+A','G-PK', 'G+A-PK','xG', 'xA', 'xG+xA', 'npxG','npxG+xA']]
df_std_stats.rename(columns={'Pos':'Position','Gls':'Goals P90','Ast':'Assist P90','G+A':'Goals + Assist P90','G-PK':'Goals - Penalty P90','G+A-PK':'Goals + Assists - Penalty P90','xG':'xG P90', 'xA':'xA P90', 'xG+xA':'xG+xA P90', 'npxG':'npxG P90','npxG+xA':'npxG+xA P90'},inplace = True)
```

```python
lst = ['Unnamed: 0_level_0','Unnamed: 1_level_0','Unnamed: 3_level_0','Unnamed: 4_level_0','Performance']
df_Card_stats = read_fbref(UrlStdStats,lst)
df_Card_stats = df_Card_stats[[ 'Player', 'Pos', 'Squad','CrdY','CrdR']]
df_Card_stats.rename(columns={'Pos':'Position','CrdY':'Yellow Card','CrdR':'Red Card'},inplace = True)
```

```python
lst = ['Unnamed: 0_level_0','Unnamed: 1_level_0','Unnamed: 3_level_0','Unnamed: 4_level_0','Unnamed: 7_level_0','Total','Unnamed: 25_level_0','Unnamed: 26_level_0','Unnamed: 27_level_0','Unnamed: 28_level_0','Unnamed: 29_level_0']
df_pass_stats = read_fbref(UrlPassStats,lst)
df_pass_stats =df_pass_stats[['Player', 'Pos', 'Squad', '90s','Cmp%','KP', '1/3', 'PPA', 'CrsPA', 'Prog']]
df_pass_stats.rename(columns={'Pos':'Position','Cmp%':'Pass Completion%','KP':'Key Passes','1/3':'Passes into Final Third','PPA':'Passes into Penalty Area','CrsPA':'Crosses into Penalty Area','Prog':'Progressive Passes'},inplace = True)
```

```python
#display(df_def)
def PassstatstoP90(row):
    row['Key Passes'] = (round(float(row['Key Passes'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Passes into Final Third'] = (round(float(row['Passes into Final Third'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Crosses into Penalty Area'] = (round(float(row['Crosses into Penalty Area'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Passes into Penalty Area'] = (round(float(row['Passes into Penalty Area'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Progressive Passes'] = (round(float(row['Progressive Passes'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    return row
df_pass_stats = df_pass_stats.apply(PassstatstoP90,axis =1)
#df_pass_stats.rename(columns={'Tackles Won':'Tackles Won P90','Dribbled Past':'Dribbled Past P90','Shots Blocked': 'Shots Blocked P90','Pass Blocked':'Pass Blocked P90','Errors led to Shot':'Errors led to Shot P90'},inplace = True)
```

```python
#display(df_def)
def DeffensivestatstoP90(row):
    row['Tackles Won'] = (round(float(row['Tackles Won'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Dribbled Past'] = (round(float(row['Dribbled Past'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Shots Blocked'] = (round(float(row['Shots Blocked'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Pass Blocked'] = (round(float(row['Pass Blocked'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Errors led to Shot'] = (round(float(row['Errors led to Shot'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    return row
df_def = df_def.apply(DeffensivestatstoP90,axis =1)
df_def.rename(columns={'Tackles Won':'Tackles Won P90','Dribbled Past':'Dribbled Past P90','Shots Blocked': 'Shots Blocked P90','Pass Blocked':'Pass Blocked P90','Errors led to Shot':'Errors led to Shot P90'},inplace = True)
```

```python
#display(df_attack)
def AttackingstatstoP90(row):
    row['Successful dribbles'] = (round(float(row['Successful dribbles'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    row['Fouls Drawn'] = (round(float(row['Fouls Drawn'])/float(row['90s']),2) if float(row['90s']) > 0 else 0)
    return row

df_attack = df_attack.apply(AttackingstatstoP90,axis =1)
df_attack.rename(columns={'Successful dribbles':'Successful dribbles P90','Fouls Drawn':'Fouls Drawn P90'},inplace = True)
```

```python
#removes the country code 3/2 letter from the squad column
import re 
def clean(col):
    col = re.sub(r'^[a-z]+\ ','',col)
    return col
df_def['Squad'] = df_def['Squad'].apply(clean)
df_attack['Squad'] = df_attack['Squad'].apply(clean)
df_std_stats['Squad'] = df_std_stats['Squad'].apply(clean)
df_pass_stats['Squad'] = df_pass_stats['Squad'].apply(clean)
df_Card_stats['Squad'] = df_Card_stats['Squad'].apply(clean)
```

```python
df_std_stats.sort_values(by=['Starts'],ascending=False,inplace = True)
df_def.sort_values(by=['90s'],ascending=False,inplace = True)
df_attack.sort_values(by=['90s'],ascending=False,inplace = True)
df_pass_stats.sort_values(by=['90s'],ascending=False,inplace = True)
```

```python
df= pd.merge(df_std_stats, df_Card_stats, on=['Player','Position','Squad'])
df= pd.merge(df, df_pass_stats, on=['Player','Position','Squad']).copy()

df = pd.merge(df, df_attack, on=['Player','Position','Squad','90s']).copy()

df = pd.merge(df, df_def, on=['Player','Position','Squad','90s']).copy()
```
